```python
from pylab import rcParams
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from heatmap import heatmap, corrplot  # Importa las bibliotecas necesarias

# Establece el tamaño de la figura
rcParams['figure.figsize'] = 7, 7

# Configura los estilos de seaborn
sns.set(color_codes=True, font_scale=1.2)

# Configuración para mostrar gráficos en línea
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2
```

```python
# Selecciona las columnas de interés
cols_of_interest = ['CH04', 'CH07', 'CH08', 'ESTADO','NIVEL_ED', 'CAT_INAC','IPCF']
subset_data = microdata_filtered[cols_of_interest]
## Transformaremos variables categóricas no binarias en dummies con categorías más simples y con información relevante para el análisis en cuestión.
# CH07: estatus civil
# Defino una función para asignar el valor booleano a las categorías 
# 1 unido o casado (tiene pareja), 0 separado divorciado, viudo o soltero (no tiene pareja). 
subset_data.loc[:, 'd_pareja'] = subset_data['CH07'].apply(lambda x: 1 if x in [1,2] else 0)

## CH08: cobertura médica. Crearemos una variable que tome el valor 0 si no paga ni le descuentan ningun seguro médico (categoría 4)
#, y que toma el valor 1 de otro modo.
subset_data.loc[:, 'd_cobertura'] = subset_data['CH08'].apply(lambda x: 0 if x in [4] else 1)

# ESTADO: Crearemos dos dummies: D_desocupado D_inactivo ú otros
subset_data.loc[:, 'd_desocupado'] = subset_data['ESTADO'].apply(lambda x: 1 if x in [2] else 0)
subset_data.loc[:, 'd_inactivo'] = subset_data['ESTADO'].apply(lambda x: 1 if x in [3, 4] else 0)
# CAT_INAC: Creamos dummy sujeto a ser inactivo. 1 si pertenece a una categoría que tiene posibilidad de volver a la actividad, 0 de otro modo.
# Las categorías consideradas como 1 son ama de casa, estudiante y rentista.
subset_data.loc[:, 'd_pos_activo'] = subset_data['CAT_INAC'].apply(lambda x: 1 if x in [2, 3, 4] else (0 if x in [1, 5, 6] else np.nan))

## Convertiré la variable de nivel educativo en una variable continua.
#NIVEL_ED
education_mapping = {
    7: 0, 
    1: 3,  
    2: 6,  
    3: 9, 
    4: 12,
    5:15,
    6:18}
subset_data['educ_years'] = subset_data['NIVEL_ED'].map(education_mapping)
# Años de educación imputa la mitad de años de educación a cada grado no culminado.
```

```python
# Eliminamos las variables originales de la base de datos
subset_data = subset_data.drop(columns=['CH07','CH08','ESTADO','NIVEL_ED','CAT_INAC'])
num_columns = subset_data.shape[1]
# Calcula la matriz de correlación
corr_matrix = subset_data.corr()
# Obtiene las etiquetas actuales
current_labels = list(corr_matrix.columns)
# Sustituye las que quieres cambiar CAMBIAR
for index, label in enumerate(current_labels):
    if label == 'CH04':
        current_labels[index] = 'Sexo'
    elif label == 'd_pareja':
        current_labels[index] = 'Tiene pareja'
    elif label == 'd_cobertura':
        current_labels[index] = 'Cobertura Médica'
    elif label == 'd_inactivo':
        current_labels[index] = 'Inactivo/a' 
    elif label == 'd_desocupado':
        current_labels[index] = 'Desocupado/a' 
    elif label == 'd_pos_activo':
        current_labels[index] = 'Posibilidad de actividad'    
    elif label == 'educ_years':
        current_labels[index] = 'Años de educación'   
# Genera el heatmap
plt.figure(figsize=(num_columns, num_columns))
#heatmap = sns.heatmap(corr_matrix, annot=True, fmt=".4f", cmap='Blues', square=True, cbar=False, linewidths=0.5)
heatmap = sns.heatmap(corr_matrix, annot=True, fmt=".4f", cmap='coolwarm', square=True, cbar=False, linewidths=0.5)

# Modifica las etiquetas para que muestren las descripciones
heatmap.set_xticklabels(current_labels, rotation=45)
heatmap.set_yticklabels(current_labels, rotation=0)
# Achica el texto dentro de los cuadros
heatmap.tick_params(axis='both', which='both', labelsize=8)

# Añade negrita y aumenta el tamaño de las etiquetas en los ejes
for tick in heatmap.get_xticklabels():
    tick.set_fontweight("bold")
    tick.set_fontsize(10)
for tick in heatmap.get_yticklabels():
    tick.set_fontweight("bold")
    tick.set_fontsize(10)
plt.savefig("Matriz_correlacion.png", dpi=300, bbox_inches='tight')

plt.show()
```

#### Inciso f
Utilizando el archivo tabla adulto equiv.xlsx, agreguen a su base de
datos una columna llamada adulto equiv que contenga los valores de
adulto equivalente de cada persona según su sexo y edad (por ejemplo,
a un varón de 2 años le corresponde 0.46). Finalmente, con el comando
groupby sumen esta nueva columna para las personas que pertenecen a un
mismo hogar y guarden ese dato en una columna llamada ad equiv hoga


```python
# Armamos una función para no tener que lidiar con el tedioso uso del archivo de Excel.

# Esta función, asignar_valor, toma dos argumentos, edad y genero, y asigna un valor numérico basado en reglas condicionales
# que dependen de estos dos argumentos. Las condiciones se prueban en orden y, cuando se encuentra una coincidencia, se
# devuelve un valor específico. Si no se cumple ninguna condición, la función devuelve None.
#Las condiciones están diseñadas para categorizar a las personas en grupos según su edad y género y asignarles un valor numérico basado en esas categorías.

def asignar_valor(edad, genero):
    if edad < 1:
        if genero == 2:
            return 0.35
        elif genero == 1:
            return 0.35
    elif edad == 1:
        if genero == 2:
            return 0.37
        elif genero == 1:
            return 0.37
    elif edad == 2:
        if genero == 2:
            return 0.46
        elif genero == 1:
            return 0.46
    elif edad == 3:
        if genero == 2:
            return 0.51
        elif genero == 1:
            return 0.51
    elif edad == 4:
        if genero == 2:
            return 0.55
        elif genero == 1:
            return 0.55
    elif edad == 5:
        if genero == 2:
            return 0.60
        elif genero == 1:
            return 0.60
    elif edad == 6:
        if genero == 2:
            return 0.64
        elif genero == 1:
            return 0.64
    elif edad == 7:
        if genero == 2:
            return 0.66
        elif genero == 1:
            return 0.66
    elif edad == 8:
        if genero == 2:
            return 0.68
        elif genero == 1:
            return 0.68
    elif edad == 9:
        if genero == 2:
            return 0.69
        elif genero == 1:
            return 0.69
    elif edad == 10:
        if genero == 2:
            return 0.70
        elif genero == 1:
            return 0.79
    elif edad == 11:
        if genero == 2:
            return 0.72
        elif genero == 1:
            return 0.82
    elif edad == 12:
        if genero == 2:
            return 0.74
        elif genero == 1:
            return 0.85
    elif edad == 13:
        if genero == 2:
            return 0.76
        elif genero == 1:
            return 0.90
    elif edad == 14:
        if genero == 2:
            return 0.76
        elif genero == 1:
            return 0.96
    elif edad == 15:
        if genero == 2:
            return 0.77
        elif genero == 1:
            return 1.00
    elif edad == 16:
        if genero == 2:
            return 0.77
        elif genero == 1:
            return 1.03
    elif edad == 17:
        if genero == 2:
            return 0.77
        elif genero == 1:
            return 1.04
    elif 18 <= edad <= 29:
        if genero == 2:
            return 0.76
        elif genero == 1:
            return 1.02
    elif 30 <= edad <= 45:
        if genero == 2:
            return 0.77
        elif genero == 1:
            return 1.00
    elif 46 <= edad <= 60:
        if genero == 2:
            return 0.76
        elif genero == 1:
            return 1.00
    elif 61 <= edad <= 75:
        if genero == 2:
            return 0.67
        elif genero == 1:
            return 0.83
    elif edad > 75:
        if genero == 2:
            return 0.63
        elif genero == 1:
            return 0.74
    else:
        return None  # En caso de que no haya una combinación válida
```

```python
# Aplicamos la funcion al data frame. Creamos una nueva columna llamada 'adulto_equiv' en el DataFrame 'microdata_filtered', donde se calculan y asignan valores equivalentes para adultos
# a partir de las edades ('CH06') y géneros ('CH04') de cada fila utilizando una función llamada 'asignar_valor'.

microdata_filtered['adulto_equiv'] = microdata_filtered.apply(lambda row: asignar_valor(row['CH06'], row['CH04']), axis=1)
```

```python
# Limpiamos y Renombramos

microdata_filtered = microdata_filtered.dropna(subset=['adulto_equiv'])

result = microdata_filtered.groupby(by=['CODUSU','NRO_HOGAR']).agg({'adulto_equiv': 'sum'}).reset_index()
result.rename(columns={'adulto_equiv': 'ad_equiv_hogar'}, inplace=True)

microdata_filtered_FINAL = pd.merge(microdata_filtered, result[['CODUSU','NRO_HOGAR', 'ad_equiv_hogar']],
                     on=['CODUSU','NRO_HOGAR'],
                     how='left')
```
